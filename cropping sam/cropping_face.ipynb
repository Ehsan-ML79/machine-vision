```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('sam.jpg')
plt.imshow(image[...,::-1])
```

```python

# to crop out the rectangle we want
cropped = image[370:550 , 440:560]

plt.imshow(cropped[...,::-1])#BGR 2 RGB
```

```python
import cv2

# Save the cropped image to a file
cv2.imwrite('cropped_sam.jpg', cropped)
```
