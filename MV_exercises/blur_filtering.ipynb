```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('sam_face2.jpg')
plt.imshow(image[...,::-1])
```

```python
height, width = image.shape[:2]

# Simply use indexing to crop out the rectangle we desire
cropped = image[0:120 ,140:240 ]

plt.imshow(cropped[...,::-1])
```

```python
# Creating our 7 x 7 kernel
kernel_7x7 = np.ones((7, 7), np.float32) / 49

blurred = cv2.filter2D(cropped, -1, kernel_7x7)

plt.figure(figsize=[15,5])
# plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Original");
# plt.subplot(132);plt.imshow(blurred[...,::-1]);plt.title("blurred with 3x3 filter");
# plt.subplot(133);
plt.imshow(blurred[...,::-1]);plt.title("blurred with 7x7 filter");
```

```python
image[0:120, 140:240] = blurred
plt.imshow(image[...,::-1]);plt.title("blurred with base image");
```
