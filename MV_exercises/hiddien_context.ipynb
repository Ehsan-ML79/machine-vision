```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

#image source: https://en.wikipedia.org/wiki/Histogram_equalization
image = cv2.imread('hidden.png', 0)
result = cv2.equalizeHist(image)

plt.figure(figsize=[12,7])
plt.subplot(221);plt.imshow(image, cmap='gray');plt.title("Original image");
plt.subplot(222);plt.hist(image.ravel(), 256);

plt.subplot(223);plt.imshow(result, cmap='gray');plt.title("equalized histogram");
plt.subplot(224);plt.hist(result.ravel(), 256);#ravel() conigous 2d array or multi array into 1d
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = cv2.imread('hidden.png', 0)

# create a CLAHE object (Arguments are optional).
clahe = cv2.createCLAHE(clipLimit=12.0, tileGridSize=(8,8))
result = clahe.apply(image)

plt.figure(figsize=[12,4])
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("After adaptive histogram equalization");

result_eq=cv2.equalizeHist(image)
plt.subplot(121);plt.imshow(result_eq, cmap='gray');plt.title("After equalization");
```
