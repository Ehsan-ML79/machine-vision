```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

gamma = 2
g12 = 1.2
g13 = 1.3
g14 = 0.4

img_original = cv2.imread("Weeki_Wachee_spring_10079u.jpg")
lookUpTable = np.empty((1,256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
outg2 = cv2.LUT(img_original, lookUpTable)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, g12) * 255.0, 0, 255)
outg12 = cv2.LUT(img_original, lookUpTable)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, g13) * 255.0, 0, 255)
outg13 = cv2.LUT(img_original, lookUpTable)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, g14) * 255.0, 0, 255)
outg14 = cv2.LUT(img_original, lookUpTable)


plt.figure(figsize=[15,5])
plt.subplot(151);plt.imshow(img_original[...,::-1]);plt.title("Original");
plt.subplot(152);plt.imshow(outg2[...,::-1]);plt.title("Output_g2");
plt.subplot(153);plt.imshow(outg12[...,::-1]);plt.title("Output_g1.2");
plt.subplot(154);plt.imshow(outg13[...,::-1]);plt.title("Output_g1.3");
plt.subplot(155);plt.imshow(outg14[...,::-1]);plt.title("Output_g1.4");
```
